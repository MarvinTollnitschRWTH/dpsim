### Load Results

```python
plt.close('all')
fig1 = plt.figure()

plt.plot(ts_dpsim_EMT['i_gen_0'].time, ts_dpsim_EMT['i_gen_0'].values, "r-", label='I_VBR_a')
plt.plot(ts_dpsim_EMT['i_gen_1'].time, ts_dpsim_EMT['i_gen_1'].values, "g-", label='I_VBR_b')
plt.plot(ts_dpsim_EMT['i_gen_2'].time, ts_dpsim_EMT['i_gen_2'].values, "b-", label='I_VBR_c')

#plt.plot(ts_dpsim_SSN_EMT['I_sg4o_SSN_0'].time, ts_dpsim_SSN_EMT['I_sg4o_SSN_0'].values, "rx", markevery=10, label='I_a_SSN')
#plt.plot(ts_dpsim_SSN_EMT['I_sg4o_SSN_1'].time, ts_dpsim_SSN_EMT['I_sg4o_SSN_1'].values, "gx", markevery=10, label='I_b_SSN')
#plt.plot(ts_dpsim_SSN_EMT['I_sg4o_SSN_2'].time, ts_dpsim_SSN_EMT['I_sg4o_SSN_2'].values, "bx", markevery=10, label='I_c_SSN')

plt.legend(loc = 4)
#plt.legend(bbox_to_anchor=(1,1))

plt.title('VBR SG model simulation: Current')
plt.xlabel('t [s]')
plt.ylabel('Phase current [A]')
```

```python
plt.close('all')
fig1 = plt.figure()

plt.plot(ts_dpsim_EMT['v_gen_0'].time, ts_dpsim_EMT['v_gen_0'].values, "r-", label='V_VBR_a')
plt.plot(ts_dpsim_EMT['v_gen_1'].time, ts_dpsim_EMT['v_gen_1'].values, "g-", label='V_VBR_b')
plt.plot(ts_dpsim_EMT['v_gen_2'].time, ts_dpsim_EMT['v_gen_2'].values, "b-", label='V_VBR_c')

#plt.plot(ts_dpsim_SSN_EMT['V_sg4o_SSN_0'].time, ts_dpsim_SSN_EMT['V_sg4o_SSN_0'].values, "rx", markevery=10, label='V_a_SSN')
#plt.plot(ts_dpsim_SSN_EMT['V_sg4o_SSN_1'].time, ts_dpsim_SSN_EMT['V_sg4o_SSN_1'].values, "gx", markevery=10, label='V_b_SSN')
#plt.plot(ts_dpsim_SSN_EMT['V_sg4o_SSN_2'].time, ts_dpsim_SSN_EMT['V_sg4o_SSN_2'].values, "bx", markevery=10, label='V_c_SSN')

plt.legend(loc = 4)
#plt.legend(bbox_to_anchor=(1,1))

plt.title('VBR SG model simulation: Voltage')
plt.xlabel('t [s]')
plt.ylabel('Phase voltage [V]')
```
